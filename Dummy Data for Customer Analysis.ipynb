```python
# Import libreries
import pandas as pd
import numpy as np
```

```python
# Create 8 random columns and 1000 rows and one more column with customer id

data = np.array([np.random.randint(low=1, high=20, size=8) for i in range(1000)])
Customer_id= np.array([np.random.randint(low=1, high=20, size=1) for i in range(1000)])
```

```python
data
```

```python
# DataFrame and name columns

df= pd.DataFrame(data, columns=['Marketing', 'Sales', 'Logistics', 'Department 1', 'Department 2', 
                                'Department 3', 'Department 4', 'Department 5'])

df['Customer'] = Customer_id
df['Customer'].astype('str')
```

```python
df.head()
```

```python
# First department target: Marketing_target. Segments by %.
s = np.random.choice([15, 12, 10, 8], p=[0.10, 0.20, 0.40, 0.30], size = (1000,1))
df1 = pd.DataFrame(s, columns=['Seg'])
df['Marketing_Target'] = df1['Seg']
```

```python
df['Segment'] = df['Marketing_Target']
```

```python
df
```

```python
# Replace number of target by segment name

df['Segment'] = df['Segment'].replace({15:'Strategic', 12:'Emerging', 10:'Core', 8:'Competitive'})
```

```python
df
```

```python
df['Segment'].unique()
```

```python
# Replace random numbers for the amount of visits

df['Marketing_Target'] = df['Customer'].replace({1:8, 2:10, 3:12, 4:15, 5:10, 6:8, 7:8, 8:10, 9:10,
                                       10:8, 11:10, 12:12, 13:15, 14:8, 15:12, 16:10, 17:8,
                                       18:12, 19:10})
```

```python
# Replace random numbers for Customer Segment

df['Segment'] = df['Customer'].replace({1:"Competitive", 2: "Core", 3: "Emerging", 4:"Strategic",
                                       5:"Core", 6:"Competitive", 7:"Competitive", 8:"Core",
                                       9:"Core", 10:"Competitive", 11:"Core", 12:"Emerging",
                                       13:"Strategic", 14: "Competitve", 15:"Emerging", 16:"Core",
                                       17:"Competitive", 18:"Emerging", 19:"Core"})
```

```python
df.columns
```

```python
# Create new columns for department targets

df['Sales_Target'] = df['Marketing_Target']
df['Logistics_Target'] = df['Marketing_Target']
df['Department 1_Target'] = df['Marketing_Target']
df['Department 2_Target'] = df['Marketing_Target']
df['Department 3_Target'] = df['Marketing_Target']
df['Department 4_Target'] = df['Marketing_Target']
df['Department 5_Target'] = df['Marketing_Target']
```

```python
# Create calculated column with the price for one more visit 

calc = (df['Marketing']-df['Marketing_Target'])*125
df['Marketing_Incremental_cost'] = calc
df['Marketing_Incremental_cost'] = df['Marketing_Incremental_cost'].mask(df['Marketing_Incremental_cost'].lt(0),0)
```

```python
calc2 = (df['Sales']-df['Sales_Target'])*125
df['Sales_Incremental_cost'] = calc2
df['Sales_Incremental_cost'] = df['Sales_Incremental_cost'].mask(df['Sales_Incremental_cost'].lt(0),0)
```

```python
calc3 = (df['Logistics']-df['Logistics_Target'])*125
df['Logistics_Incremental_cost'] = calc3
df['Logistics_Incremental_cost'] = df['Logistics_Incremental_cost'].mask(df['Logistics_Incremental_cost'].lt(0),0)
```

```python
calc4 = (df['Department 1']-df['Department 1_Target'])*125
df['Department 1_Incremental_cost'] = calc4
df['Department 1_Incremental_cost'] = df['Department 1_Incremental_cost'].mask(df['Department 1_Incremental_cost'].lt(0),0)
```

```python
calc5 = (df['Department 2']-df['Department 2_Target'])*125
df['Department 2_Incremental_cost'] = calc5
df['Department 2_Incremental_cost'] = df['Department 2_Incremental_cost'].mask(df['Department 2_Incremental_cost'].lt(0),0)
```

```python
calc6 = (df['Department 3']-df['Department 3_Target'])*125
df['Department 3_Incremental_cost'] = calc6
df['Department 3_Incremental_cost'] = df['Department 3_Incremental_cost'].mask(df['Department 3_Incremental_cost'].lt(0),0)
```

```python
calc7 = (df['Department 4']-df['Department 4_Target'])*125
df['Department 4_Incremental_cost'] = calc7
df['Department 4_Incremental_cost'] = df['Department 4_Incremental_cost'].mask(df['Department 4_Incremental_cost'].lt(0),0)
```

```python
calc8 = (df['Department 5']-df['Department 5_Target'])*125
df['Department 5_Incremental_cost'] = calc8
df['Department 5_Incremental_cost'] = df['Department 5_Incremental_cost'].mask(df['Department 5_Incremental_cost'].lt(0),0)
```
